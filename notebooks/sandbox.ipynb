```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

a = np.random.randn(1000)
b = np.random.randn(1000)
c = np.random.choice([0, 1], size=1000)
d = ["Other data"] * 1000
df = pd.DataFrame({"a": a, "b": b, "c": c, "d": d})
sns.scatterplot(data=df, x="a", y="b", hue="c", style="d", markers={"Other data": "s"})
plt.show()
```

```python
import pymc as pm

coords = {"bes":[1,2,3]}
with pm.Model(coords=coords) as model:
    d = pm.Dirichlet("d", a=[1,1,1], dims="bes")
    print(d.shape.eval())
```

```python
import pymc as pm

coords = {"bes":[1,2,3]}
with pm.Model(coords=coords) as model:
    d = pm.StickBreakingWeights("d", alpha=1, K=2)
    print(d.eval())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pymc as pm

g = pm.Gamma.dist(1, 0.5)
lin = np.linspace(-0.5, 10, 200)
y = pm.logp(g, lin).exp().eval()
plt.plot(lin, y)
```

```python
from pcntoolkit.math_functions.basis_function import (
    BsplineBasisFunction,
    PolynomialBasisFunction,
)

x = np.random.randn(200)
x.sort()
x=x[:, None]
x = np.linspace(-2,2,200)[:,None]
b = BsplineBasisFunction(0, 3, 6)
# b = PolynomialBasisFunction(0, 5)
b.fit(x)
Phi = b.transform(x)
for f in range(Phi.shape[1]):
    plt.plot(x, Phi[:,f], label = f)
plt.legend()
```
